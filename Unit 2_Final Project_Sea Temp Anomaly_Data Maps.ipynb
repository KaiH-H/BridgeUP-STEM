```python
# Imports the needed libraries
import matplotlib.pyplot as plt 
import numpy as np
import csv
```

```python
csvFile = open("sea_temp_anom_Dec_2002.csv", 'r') # r means read; no changes can be made to the file
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
print(len(data_lists[0]))
```

```python
print(len(data_lists))
```

```python
width = len(data_lists[0])
height = len(data_lists)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8) 
```

```python
all_values = [] 

for r, row in enumerate(data_lists):
    for c, value in enumerate(row): 
        if float(value) < 99998:
            all_values.append(float(value)) 
        if float(value) <= -1:
            grid[r, c] = [0, 33, 91] #Dark Blue
        elif float(value) <= -0.5:
            grid[r, c] = [0, 64, 178] # Medium Blue
        elif float(value) <= .2:
            grid[r, c] = [0, 80, 214] # Medium-regular blue
        elif float(value) <= .5:
            grid[r, c] = [2, 95, 255] # Regular blue
        elif float(value) <= 1:
            grid[r, c] = [28, 110, 255] # Lighter blue
        elif float(value) <= 1.5:
            grid[r, c] = [43, 119, 255]
        elif float(value) <= 2.5:
            grid[r, c] = [73, 138, 255]
        elif float(value) <= 3.5:
            grid[r, c] = [102, 155, 255]
        elif float(value) <= 5:
            grid[r, c] = [135, 176, 255]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(max(all_values))
print(min(all_values))
```

```python
plt.title("Global Ocean Temperature in Celsius (2002)") 
plt.imshow(grid) 
plt.axis('off') 
plt.savefig("Ocean_Temp_Anom_2002.png",dpi = 1000)
plt.show()
```
