```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SeaTemp_SeaLevel_anomaly.csv")

df.head()
```

```python
year = df["Year"]
temp = df["Sea temperature anomaly (C)"]
mm = df["Change in mean sea level (mm)"]
```

```python
print(max(year))
print(min(year))
```

```python
plt.plot(year, temp, label = "Temp. Anomaly", c = "r")
plt.xlabel("Year")
plt.ylabel("Sea Temperature Anomaly (C)")
plt.legend(loc = 2)
plt.twinx()
plt.plot(year, mm, label = "Change in ocean level", c = "blue")
plt.xlabel("Year")
plt.ylabel("Change in mean ocean level (mm)")
plt.legend(loc = 1)
plt.title("Change in Temperature and Ocean Level from 1992 to 2016")
plt.savefig("Changes in Ocean Temp and Water Level.png", dpi = 1000)
plt.show()
```
